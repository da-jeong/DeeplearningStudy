```python
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean, y_mean)
```

```python
denominator = sum([(j-x_mean)**2 for j in x])

def top(x, x_mean, y, y_mean):
  d = 0
  for j in range(len(x)):
    d += (x[j] - x_mean) * (y[j] - y_mean)
  return d

numerator = top(x, x_mean, y, y_mean)

a = numerator / denominator
b = y_mean - a * x_mean

print(a, b)
```

```python
def predict(x):
  fake_a = 3
  fake_b = 76
  return fake_a * x + fake_b

y_pred = []

for j in range(len(x)):
  y_pred.append(predict(x[j]))

def mse(y, y_pred):
  return (1/len(y)) * sum((y-y_pred) ** 2)

print("mse:" + str(mse(y, y_pred)))
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection="3d")
ax.scatter3D(x1, x2, y, s=200)
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.01
epoch = 2001
n= len(x1)

for j in range(epoch):
  y_pred = a1 * x1 + a2 * x2 + b
  error = y - y_pred

  a1_diff = (2/n) * sum(-x1 * error)
  a2_diff = (2/n) * sum(-x2 * error)
  b_diff = (2/n) * sum(-error)

  a1 = a1 - lr * a1_diff
  a2 = a2 - lr * a2_diff
  b = b - lr * b_diff

  if (j % 100) == 0:
    print("epochhhh = %.f, 1 = %.4f 2=%.4f y=%.4f" %(j, a1, a2, b))
  print(y)
  print(y_pred)
```

```python
from tensorflow import keras
from keras import Sequential, Input
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt


x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(1,activation="linear"))

model.compile(optimizer="sgd", loss="mse")

model.fit(x, y, epochs=2000)

y_pred = model.predict(x)

plt.figure()
plt.scatter(x, y, marker='o', c='k', s=100)
plt.plot(x, y_pred, marker='*', c='r', markersize=15)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

hour = 7
prediction = model.predict(np.array([hour]))

print("%.f     %.2f" %(hour, prediction))
```
